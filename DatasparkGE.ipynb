```python
import pandas as pd
from datetime import datetime
```

```python
customersdf=pd.read_csv("C:/Users/hp/Desktop/DataSpark_Project/Customers.csv",encoding="ISO-8859-1")
exchange_ratesdf=pd.read_csv("C:/Users/hp/Desktop/DataSpark_Project/Exchange_Rates.csv",encoding="ISO-8859-1")
productsdf=pd.read_csv("C:/Users/hp/Desktop/DataSpark_Project/Products.csv",encoding="ISO-8859-1")
storesdf=pd.read_csv("C:/Users/hp/Desktop/DataSpark_Project/Stores.csv",encoding="ISO-8859-1")
salesdf=pd.read_csv("C:/Users/hp/Desktop/DataSpark_Project/Sales.csv",encoding="ISO-8859-1")
```

```python
exchange_ratesdf=pd.read_csv("C:/Users/hp/Desktop/DataSpark_Project/Exchange_Rates.csv",encoding="ISO-8859-1")
```

```python
customersdf["Birthday"]=pd.to_datetime(customersdf["Birthday"])
```

```python
customersdf["Birthday"]=customersdf["Birthday"].dt.strftime('%Y-%m-%d')
```

```python
customersdf.rename(columns={'State Code':'State_Code'},inplace=True)
```

```python
customersdf.rename(columns={'Zip Code':'Zip_Code'},inplace=True)
```

```python
customersdf
```

```python
exchange_ratesdf["Date"]=pd.to_datetime(exchange_ratesdf["Date"])
exchange_ratesdf["Date"]=exchange_ratesdf["Date"].dt.strftime('%Y-%m-%d')
```

```python
exchange_ratesdf.rename(columns={'Date':'exchange_date'},inplace=True)
```

```python
exchange_ratesdf
```

```python
productsdf["Unit Cost USD"]=productsdf["Unit Cost USD"].apply(lambda i:str(i.replace(",","")))
productsdf["Unit Price USD"]=productsdf["Unit Price USD"].apply(lambda i:str(i.replace(",","")))
```

```python
productsdf["Unit Cost USD"]=productsdf["Unit Cost USD"].apply(lambda i:str(i.replace("$","")))
productsdf["Unit Price USD"]=productsdf["Unit Price USD"].apply(lambda i:str(i.replace("$","")))
```

```python
productsdf.rename(columns={'Product Name':'Product_Name'},inplace=True)
productsdf.rename(columns={'Unit Cost USD':'Unit_Cost_USD'},inplace=True)
productsdf.rename(columns={'Unit Price USD':'Unit_price_USD'},inplace=True)
```

```python
productsdf
```

```python
salesdf["Order Date"]=pd.to_datetime(salesdf["Order Date"])
salesdf["Order Date"]=salesdf["Order Date"].dt.strftime('%Y-%m-%d')
salesdf["Delivery Date"]=pd.to_datetime(salesdf["Delivery Date"])
salesdf["Delivery Date"]=salesdf["Delivery Date"].dt.strftime('%Y-%m-%d')
```

```python
salesdf["Delivery Date"].fillna(value='9999-12-31',inplace=True)
```

```python
salesdf["Delivery Date"]=salesdf["Delivery Date"].astype(str)
```

```python
salesdf.rename(columns={'Order Number':'Order_Number'},inplace=True)
salesdf.rename(columns={'Line Item':'Line_Item'},inplace=True)
salesdf.rename(columns={'Order Date':'Order_Date'},inplace=True)
salesdf.rename(columns={'Delivery Date':'Delivery_Date'},inplace=True)
salesdf.rename(columns={'Currency Code':'Currency_Code'},inplace=True)
```

```python
salesdf
```

```python
storesdf["Open Date"]=pd.to_datetime(storesdf["Open Date"])
storesdf["Open Date"]=storesdf["Open Date"].dt.strftime('%Y-%m-%d')
```

```python
storesdf["Square Meters"]=storesdf["Square Meters"].fillna(0)
```

```python
storesdf.rename(columns={'Square Meters':'Square_Meters'},inplace=True)
storesdf.rename(columns={'Open Date':'Open_Date'},inplace=True)
```
